```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report
from pydub import AudioSegment
from pydub.playback import play
```

```python
df = pd.read_csv("features_30_sec.csv")
labels=df['label']
num=0
one_hot_encode={}
labels_encoded=[]

for i in labels:
    if i not in one_hot_encode.keys():
        one_hot_encode[i]=num
        labels_encoded.append(num)
        num+=1
    else:
        labels_encoded.append(one_hot_encode[i])
df['labels_encoded']=labels_encoded
print("Encoding of the music genre classes:",one_hot_encode)
df
```
